# ðŸ”§ TransformaciÃ³n

Usamos df.drop para poder eliminar las columnas antiguas y ya junto a concat podemos unir las nuevas columnas ya utilizando normalize

```python
df_normalized = pd.concat([df.drop(['customer', 'phone', 'internet', 'account'], axis=1), df_customer, df_phone, df_internet, df_account], axis=1)
df_normalized
```

```python
df_normalized.info()
```

```python
unique_churn_values = df_normalized['Charges.Monthly'].unique()

print(unique_churn_values)
```

```python
df_normalized['Charges.Total'] = df_normalized['Charges.Total'].replace(r'^\s*$', np.nan, regex=True)
df_normalized['Charges.Total'] = df_normalized['Charges.Total'].astype(np.float64)
```

## Columna de cuentas diarias

```python
df_normalized['Charges.Daily'] = df_normalized['Charges.Monthly'] / 30
```

```python
pd.set_option('display.max_columns', None)
display(df_normalized)
```

## EstandarizaciÃ³n y transformaciÃ³n de datos (opcional)

```python
columnas_binarias = ['Churn','Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling']
df_normalized[columnas_binarias] = df_normalized[columnas_binarias].replace({'Yes': 1, 'No': 0})
```

```python
df_normalized
```

```python
df_normalized = df_normalized.replace({'No internet service': 0, 'No phone service': 0})
df_normalized['Churn'] = df_normalized['Churn'].replace('', 0)
```

```python
for x in columnas_binarias:
  unique_values = df_normalized[x].unique()
  print(f"Unique values in '{x}': {unique_values}")
```
